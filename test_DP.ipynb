```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('dataset.csv')
df
```

```python
df['produit']= df['Category'].str.split("/").str.get(-1)
df1=df['Category'].str.split('/',expand=True)
df1.columns = ["Category","sous catgory","type","sous type","ithem","ll"]
df1
```

```python
df1.drop('Category',inplace=True, axis=1 )
df = pd.concat([df, df1], axis=1)
df
```

```python
df2=df['Discounted Price'].str.split(' ',expand=True)
df3=df['Old Price'].str.split(' ',expand=True)
prices = pd.concat([df2, df3], axis=1)
prices.columns = ["Discounted-Price","1","2","3","4","Old-Price","5","6","7","8"]
prices=prices['Discounted-Price'].str.replace(',','')
prices=pd.to_numeric(prices)
```

```python
df= pd.concat([df, prices], axis=1)
```

```python
df3=df['Star Rating'].str.split('out',expand=True)
df3.columns = ["rate","1"]
df3=df3["rate"].str.replace('No star rating','0')
df3
```

```python
df3=pd.to_numeric(df3)
df= pd.concat([df, df3], axis=1)
df
```

```python
df.dtypes
```

```python
df['Price range'] = df['Discounted-Price'].apply(lambda x: "0-50" if (x > 0 and x < 50) else "51-100" if (x > 51 and x < 100) else "101-500" if (x > 101 and x < 500) 
else "501-1000" if (x > 501 and x < 1000) else "+1000")
```

```python
df.drop('Category',inplace=True, axis=1 )
df.drop('Link',inplace=True, axis=1 )
df.drop('Star Rating',inplace=True, axis=1 )
df.drop('Discounted Price',inplace=True, axis=1 )
df.drop('Unnamed: 0',inplace=True, axis=1 )
```

```python
cc=df['Brand'].unique()
rateavr=[]
pravr=[]
for RAW in cc :
    l=df.loc[df['Brand'] == RAW ]['rate'].sum()/df.loc[df['Brand'] == RAW ]['rate'].__len__()
    rateavr.append(l)

for RAW in cc :
    l=df.loc[df['Brand'] == RAW ]['Discounted-Price'].sum()/df.loc[df['Brand'] == RAW ]['Discounted-Price'].__len__()
    pravr.append(l)
```

```python
ccc=pd.DataFrame(cc,columns=["Brand"])
pravr=pd.DataFrame(pravr,columns=["avr price"])
rateavr=pd.DataFrame(rateavr,columns=["avr rate"])
```
